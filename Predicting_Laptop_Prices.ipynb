```python
import pandas as pd
df = pd.read_excel('/content/laptop_prices_dataset.xlsx', engine='openpyxl')
```

```python
#initial exploration

df.head()           # Preview data
df.info()           # Check data types and nulls
df.describe()       # Summary statistics
df.columns          # List column names
```

```python
# handling missing values
df.dropna()                     # Drop rows with any missing values
df.dropna(axis=1)              # Drop columns with missing values
```

```python
# filling missing values

df.fillna(0)                   # Fill with zero
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].mean(), inplace=True)
```

```python
# removing duplicates
df.drop_duplicates(inplace=True)
```

```python
# fix datatypes
df['Price_euros'] = pd.to_numeric(df['Price_euros'], errors='coerce')
```

```python
# trim and clean
df['Company'] = df['Company'].str.strip().str.lower()
```

```python
# trim and clean
df['Product'] = df['Product'].str.strip().str.lower()
```

```python
# trim and clean
df['TypeName'] = df['TypeName'].str.strip().str.lower()
```

```python
# trim and clean
df['Touchscreen'] = df['Touchscreen'].str.strip().str.lower()
```

```python
# trim and clean
df['Screen'] = df['Screen'].str.strip().str.lower()
```

```python
# trim and clean
df['IPSpanel'] = df['IPSpanel'].str.strip().str.lower()
```

```python
# trim and clean
df['RetinaDisplay'] = df['RetinaDisplay'].str.strip().str.lower()
```

```python
# trim and clean
df['CPU_company'] = df['CPU_company'].str.strip().str.lower()
```

```python
# trim and clean
df['PrimaryStorageType'] = df['PrimaryStorageType'].str.strip().str.lower()
```

```python
# Trim and clean
df['SecondaryStorageType'] = df['SecondaryStorageType'].str.strip().str.lower()
```

```python
# Trim and clean
df['GPU_company'] = df['GPU_company'].str.strip().str.lower()
```

```python
# Filter Outliers Using IQR (Numeric Columns Only)

# Select only numeric columns
numeric_cols = df.select_dtypes(include='number').columns

# Calculate IQR for each numeric column
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Filter out rows with outliers
df = df[~((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df.isnull().sum()
```

```python
# Step 1: Convert binary categorical columns to 0/1
binary_cols = ['Touchscreen', 'IPSpanel', 'RetinaDisplay']
for col in binary_cols:
    if col in df.columns:
        df[col] = df[col].apply(lambda x: 1 if str(x).strip().lower() in ['yes', 'true', '1'] else 0)

# Step 2: Encode categorical variables using one-hot encoding
categorical_cols = [
    'Company', 'TypeName', 'OS', 'CPU_company', 'CPU_model',
    'PrimaryStorageType', 'SecondaryStorageType', 'GPU_company', 'GPU_model'
]
existing_categorical_cols = [col for col in categorical_cols if col in df.columns]
df = pd.get_dummies(df, columns=existing_categorical_cols, drop_first=True)

# Step 3: Feature engineering
if 'ScreenW' in df.columns and 'ScreenH' in df.columns:
    df['Resolution'] = df['ScreenW'] * df['ScreenH']
if 'SecondaryStorage' in df.columns:
    df['Has_SecondaryStorage'] = df['SecondaryStorage'].apply(lambda x: 1 if x > 0 else 0)

# Step 4: Drop irrelevant or high-cardinality columns
drop_cols = ['Product', 'Screen']
df.drop(columns=[col for col in drop_cols if col in df.columns], inplace=True)

# Step 5: Final check
print(" Cleaned DataFrame shape:", df.shape)
print(" Remaining columns:", df.columns.tolist())
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load cleaned, encoded data
df = pd.read_excel("laptop_data_cleaned.xlsx")

# Split features and target
X = df.drop(columns=["Price_euros"])
y = df["Price_euros"]

# Train model
model = LinearRegression()
model.fit(X, y)

# --- User Input ---
brand = "LENOVO"              # e.g., "hp", "lenovo", "acer"
model_name = "Yoga 900-13ISK"     # e.g., "macbook_air", "xps_13", "ideapad"
ram = 16
primary_storage = 512
secondary_storage = 0
inches = 13.3

# --- Build input row ---
input_dict = {col: 0 for col in X.columns}  # Start with all zeros
input_dict[f"Company_{brand.lower()}"] = 1
input_dict[f"Product_{model_name.lower()}"] = 1
input_dict["Ram"] = ram
input_dict["PrimaryStorage"] = primary_storage
input_dict["SecondaryStorage"] = secondary_storage
input_dict["Inches"] = inches

# Convert to DataFrame
input_df = pd.DataFrame([input_dict])
input_df = input_df[X.columns]  # Ensure column order matches training data

# Predict
predicted_price = model.predict(input_df)[0]
print(f"\n💻 Brand: {brand.title()} | Model: {model_name.replace('_', ' ').title()}")
print(f"📏 RAM: {ram} GB | 💾 Storage: {primary_storage} + {secondary_storage} GB | 🖥 screen: {inches}″")
print(f"🎯 Predicted Price: €{predicted_price:.2f}")
```

```python
import pandas as pd                                  # for data handling
from sklearn.linear_model import LinearRegression    # for model traing
import ipywidgets as widgets                         # for UI component
from IPython.display import display, Markdown        # for rendering markdown

# Load cleaned, encoded data
df = pd.read_excel("laptop_data_cleaned.xlsx")

# Split features and target
X = df.drop(columns=["Price_euros"])
y = df["Price_euros"]

# Train model
model = LinearRegression()
model.fit(X, y)

# --- Input Widgets ---
brand_input = widgets.Text(value="lenovo", description="Brand:")
model_input = widgets.Text(value="Yoga 900-13ISK", description="Model:")
ram_input = widgets.IntText(value=16, description="RAM (GB):")
primary_input = widgets.IntText(value=512, description="Primary (GB):")
secondary_input = widgets.IntText(value=0, description="Secondary (GB):")
inches_input = widgets.FloatText(value=13.3, description="Screen (in):")
resolution_input = widgets.IntText(value=1920*1080, description="Resolution:")
touch_input = widgets.Checkbox(value=True, description="Touchscreen")
ips_input = widgets.Checkbox(value=True, description="IPS Panel")
retina_input = widgets.Checkbox(value=False, description="Retina Display")
predict_button = widgets.Button(description="Predict Price 💸", button_style='success')
output = widgets.Output()

# --- Prediction Logic ---
def on_click(b):
    with output:
        output.clear_output()
        brand = brand_input.value.strip().lower()
        model_name = model_input.value.strip().lower()
        ram = ram_input.value
        primary_storage = primary_input.value
        secondary_storage = secondary_input.value
        inches = inches_input.value
        resolution = resolution_input.value
        touchscreen = int(touch_input.value)
        ips_panel = int(ips_input.value)
        retina_display = int(retina_input.value)

        input_dict = {col: 0 for col in X.columns}
        input_dict[f"Company_{brand}"] = 1
        input_dict[f"Product_{model_name}"] = 1
        input_dict["Ram"] = ram
        input_dict["PrimaryStorage"] = primary_storage
        input_dict["SecondaryStorage"] = secondary_storage
        input_dict["Inches"] = inches
        input_dict["Resolution"] = resolution
        input_dict["Touchscreen"] = touchscreen
        input_dict["IPSpanel"] = ips_panel
        input_dict["RetinaDisplay"] = retina_display

        # Optional: If Has_SecondaryStorage is in your feature set
        if "Has_SecondaryStorage" in X.columns:
            input_dict["Has_SecondaryStorage"] = int(secondary_storage > 0)

        input_df = pd.DataFrame([input_dict])[X.columns]
        predicted_price = model.predict(input_df)[0]

        display(Markdown(f"""
### 💻 Brand: `{brand.title()}` | Model: `{model_name.title()}`
📏 RAM: `{ram} GB` | 💾 Storage: `{primary_storage} + {secondary_storage} GB` | 🖥 Screen: `{inches}″`
🧮 Resolution: `{resolution}` | 🖐 Touchscreen: `{bool(touchscreen)}` | 🎨 IPS: `{bool(ips_panel)}` | 🍎 Retina: `{bool(retina_display)}`
### 🎯 **Predicted Price: €{predicted_price:.2f}**
"""))

predict_button.on_click(on_click)

# --- Display Interface ---
display(widgets.VBox([
    brand_input, model_input, ram_input,
    primary_input, secondary_input, inches_input,
    resolution_input, touch_input, ips_input, retina_input,
    predict_button, output
]))
```
